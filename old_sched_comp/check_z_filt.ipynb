```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('baseline2018a.db')
filtername = 'z'
df = pd.read_sql('select observationStartMJD from summaryallprops where filter="%s"' % filtername, conn)
```

```python
conn = sqlite3.connect('baseline_1exp_pairsmix_10yrs.db')
df2 = pd.read_sql('select observationStartMJD from summaryallprops where filter="%s"'% filtername, conn)
```

```python
plt.plot(df['observationStartMJD']-df['observationStartMJD'].min(), np.arange(df.size)/df.size, label='proposal')
plt.plot(df2['observationStartMJD']-df2['observationStartMJD'].min(), np.arange(df2.size)/df2.size, label='FBS')
plt.plot([0,365.25*10], [0,1], '--k', alpha=.8)
plt.xlabel('Days into survey')
plt.ylabel('Fraction of %s band taken' % filtername)
plt.legend()
```
