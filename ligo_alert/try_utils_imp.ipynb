```python
import numpy as np
import healpy as hp
import matplotlib.pylab as plt
from astropy.io import fits
from lsst.sims.utils import moc2array
```

```python
hdu1 = fits.open('bayestar.fits')
```

```python
degraded_map = moc2array(hdu1[1].data['PROBDENSITY'], hdu1[1].data['UNIQ'])
```

```python
hp.mollview(degraded_map)
```

```python
degraded_map = moc2array(hdu1[1].data['PROBDENSITY'], hdu1[1].data['UNIQ'], nside=64)
hp.mollview(degraded_map)
```

```python
degraded_map = moc2array(hdu1[1].data['PROBDENSITY'], hdu1[1].data['UNIQ'], nside=64, fillVal=np.nan)
hp.mollview(degraded_map)
```

```python
degraded_map = moc2array(hdu1[1].data['PROBDENSITY'], hdu1[1].data['UNIQ'], nside=64, fillVal=hp.UNSEEN)
hp.mollview(degraded_map)
```
